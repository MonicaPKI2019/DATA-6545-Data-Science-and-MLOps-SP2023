````
# Forward loop and a backward loop 
# "\" is to create/move things to new line. \
# Need '\\' to represent the '\' symbol

# symbols to print for output initialized 
symbol_list = ['\\','=','/']


def q1(symbol_list, num_repeat ):  # function header on this line, function name q1a, two parameters (function inputs that are variables), ends with ':'
    # function body always indented 
    # First define a forward loop, iterate through 5 times forward 
    for x in range(num_repeat): 
        print(symbol_list[1]*x, symbol_list[0]) # print the second element of 'symbol_list', index 1 ("=") times iteration number (0 through 4); then print first element index 0 ('\')
    # Second define a reverse loop, iterate through 5 times backward 
    for x in reversed(range(num_repeat)):
        print(symbol_list[1]*x, symbol_list[2]) # print the second element of 'symbol_list', index 1 ("=") times iteration number (4 through 0); then print third element index 2 ('/') 

q1(symbol_list, 5)  # call the function and set number_repeat to 5 (so we iterate through 5 times each forward and backward loop)
````

````
# Test if void or fruitful function 
# is a void function b/c q1 does some action (print a drawing) 
# but does not have a return value (week 3 functions BA505)
# with fruitful functions we always do something about the function 
# results, like store it as a variable
resultsq1 = q1(symbol_list, 5)
print(resultsq1)
````

````
# For loop number reversal 
def num_desc(): # function header, function name 'num_desc'
    number_list = [1, 2, 3, 4, 5] 
    # for loop to iterate through the collection of elements in list 'number_list'
    for num in number_list[::-1]: # slicing, start at first element, end at last element, step of -1 to indicate to work in reverse
    # for each iteration print element in 'number_list' and concatenate the string '...' 
        print(str(num) + '...') # used string operator '+' to remove the space between the number and string '...' to produce same output as above
    # when collection 'number_list' is exhausted print out the string 'Boom!!!'
    print('Boom!!!')

# Call out function to run it
num_desc()
````

````
#alterative syntax with a function creation
first_names = ['Anna', 'Carrie', 'Justin', 'Lacie']
last_names = ['Case', 'Conda', 'Curtains', 'Oakey']

combined = [] # Empty list for last, first names
               
for last_name in last_names:    # iterate through last names list
    names = []   # create an empty list for name combinations
    for first_name in first_names:  # iterate through first names list
        names.append([last_name,first_name])  #add to the empty list 'names' by element in last name, then first name
    combined.append(names)  # add the list 'names' to the empty list 'combined'

print(combined)
````

## Q6: Substring Inserter

Write a function (`char_ins()`) to insert any user defined substring into the $n^{th}$ position of another string.

```python
char_ins('i', 3, 'python') --> 'pytihon'
char_ins('i', -287, 'python') --> 'ipython'
char_ins('i', 10000, 'python') --> 'pythoni'
char_ins('i', 2, '') --> 'i'
```

````
### Write your function here
# function has three inputs
# 'sub_str' is the string to insert into change_str
# 'n_pos' is the index number/ position number of where to insert 'sub_str' into 'change_str'
# 'change_str' is the specified string that will be modified 
def char_ins(sub_str, n_pos, change_str):
    # 'change_str' slice up to but not including the nth position
    # 'sub_str' concatenate to the sliced string 'change_str'
    # 'change_str' slice starting at nth position
    # concatenate the sliced string after nth position after 'sub_str'
    x = change_str[:n_pos] + sub_str + change_str[n_pos:] 
    print(x)
````
